# Prueba 5: Conversión de archivos tipo binario (IGOR) a txt sin utilizar el sofware de IGOR

1. Se define función para convertir de txt a csv.

```python
def conversor(archivo_txt, nombre_csv):

    import csv

    archivo = open(archivo_txt)
    lineas = archivo.readlines()
    archivo.close()

    nombre_archivo_csv = nombre_csv
    with open(nombre_archivo_csv, 'w', newline='') as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        escritor_csv.writerow(['-', 'tiempo', 'voltaje'])
        for linea in lineas:
            # Dividir cada línea en base al delimitador (en este caso, una coma)
            datos = linea.strip().split(',')
            escritor_csv.writerow(datos)
    
    return nombre_archivo_csv
```

2. Se utiliza librería creada para convertir archivos tipo binario (IGOR) a txt en Phython

```python
from igor.binarywave import load

# Ruta del archivo IGOR Binary Wave
igor_file_path = "E1.ibw"
txt_file_path = "E1_wave_data.txt"

try:
    # Cargar los datos del archivo IGOR Binary Wave
    data = load(igor_file_path)
    print(data)
    wave_data = data['wave']['wData']  # Datos de la onda
    wave_header = data['wave']['wave_header']  # Cabecera de la onda

    # Verificar si hay notas en la onda
    wave_notes = data['wave'].get('wave_note', None)

    # Guardar los datos en un archivo .txt
    with open(txt_file_path, "w", encoding="utf-8") as txt_file:
        # Escribir cabecera
        txt_file.write("Wave Header:\n")
        txt_file.write(str(wave_header) + "\n\n")
        
        # Escribir notas (si existen)
        if wave_notes:
            txt_file.write("Wave Notes:\n")
            txt_file.write(wave_notes.decode("utf-8") + "\n\n")

        # Escribir los datos en formato separado por comas
        txt_file.write("Wave Data:\n")
        if wave_data.ndim == 1:  # Datos unidimensionales
            txt_file.write(",".join(map(str, wave_data)) + "\n")
        else:  # Datos multidimensionales
            for row in wave_data:
                txt_file.write(",".join(map(str, row)) + "\n")

    print(f"Archivo convertido con éxito. Datos guardados en {txt_file_path}")
except Exception as e:
    print(f"Error al convertir el archivo: {e}")
```

3. Se aplica función conversor, para convertir el archivo txt en csv

```python
conversor(txt_file_path, "E1_wave_data.csv")
```
